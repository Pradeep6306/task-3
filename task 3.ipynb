```python
import zipfile
import os

# Define the path to the zip file and the extraction directory
zip_file_path = 'archive (1).zip'  # Update this path if necessary
extraction_directory = 'extracted_files'

# Create the extraction directory if it does not exist
os.makedirs(extraction_directory, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extraction_directory)

print(f"Files have been extracted to: {os.path.abspath(extraction_directory)}")

# List the extracted files
extracted_files = os.listdir(extraction_directory)
print("Extracted Files:", extracted_files)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'extracted_files\Iris.csv'
data = pd.read_csv(file_path)

# Create a histogram for the SepalLengthCm column
plt.figure(figsize=(8, 6))
plt.hist(data['SepalLengthCm'], bins=15, color='skyblue', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'extracted_files\Iris.csv'
data = pd.read_csv(file_path)

# Define numerical columns
numerical_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Create histograms for numerical columns
plt.figure(figsize=(12, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    plt.hist(data[column], bins=20, color='blue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Create a bar chart for the Species column
species_counts = data['Species'].value_counts()

plt.figure(figsize=(8, 6))
species_counts.plot(kind='bar', color='green', edgecolor='black')
plt.title('Distribution of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
